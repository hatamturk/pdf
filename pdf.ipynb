a) Rasgele değişkeninin dağılım fonksiyonunu bulunuz.

b) Dağılım fonksiyonu yardımıyla aşağıdaki olasılıkları bulunuz.

P1(0.20 < X <= 0.80) , P2( X <= 0.75) , P3( X > 0.35)
 
 Çözüm: 

 a ) Rasgele değişkenin dağılım fonksiyonu , olasılık yoğunluk fonksiyonu kullanılarak aşağıdaki gibi bulunmuştur.
 
denklem2.JPG


```python
def pdf(deger):
   if deger < 0:
    sonuc=0
   elif deger >=0  and deger <=1 :
    sonuc = deger * deger
   else:
    sonuc = 1      
   return sonuc
print("P1 (P1(0.20 < X <= 0.80)) Olasılığının Sonucu  :")
print(pdf(0.80) - pdf(0.20))
print("P2 (P2( X <= 0.75)) Olasılığının Sonucu  :")
print(pdf(0.75))
print("P3 (P3( X > 0.35)) Olasılığının Sonucu  :")
print(1 - pdf(0.35))
```
